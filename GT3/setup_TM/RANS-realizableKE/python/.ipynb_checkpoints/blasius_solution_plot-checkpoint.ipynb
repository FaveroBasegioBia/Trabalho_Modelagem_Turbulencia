```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load input files

data0 = np.loadtxt("../postProcessing/sets/1000/profile0_U_wallShearStress.xy",skiprows=0)
data1 = np.loadtxt("../postProcessing/sets/1000/profile1_U_wallShearStress.xy",skiprows=0)

#data extracted from paraview
data2 = np.loadtxt("./ws.csv",skiprows=1,delimiter=',')
```

```python
#Declare values and compute eta for blasius solution

nu=0.00002
rho=1.
xloc=1.8
U=1

#Compute reynolds in the sampling postion xloc
rex = U*xloc/nu

#Similarity parameter of blasius solution
#eta = data0[:,0]*np.sqrt(data0[:,2]/(nu*xloc))
eta = data0[:,0]*np.sqrt(U/(nu*xloc))


#Compute total wall shear stresses and friction coefficient
ws = np.sqrt(data1[:,4]**2 + data1[:,5]**2 + data1[:,6]**2)
cf_of=ws/(0.5*rho*U**2)
```

```python
#Compute blasius solution using shooting method
#https://github.com/numerical-mooc/assignment-bank/blob/master/Lessons.and.Assignments/Blasius.Shooting.Method/01_shooting_blasius.ipynb

nfinal = 8.  # final value of eta, called n here
dn = 0.01  # step size
N = int(nfinal/dn)  # number of time steps
n = np.linspace(0.0,nfinal,N)

# arrays to be filled with values
f = np.zeros(N)
f1 = np.zeros(N)
f2 = np.zeros(N)

# initial conditions
f[0] = 0. 
f1[0] = 0.

# our shot for the value of f at infinity
f2shot = np.linspace(0.0,5.0,1000)

for i in range(len(f2shot)):
    f2[0] = f2shot[i]
    
    # iterating using euler's method
    for j in range(0,N-1):
        f[j+1] = f[j] + f1[j]*dn
        f1[j+1] = f1[j] + f2[j]*dn
        f2[j+1] = f2[j] - 0.5*f[j]*f2[j]*dn
    #if f1[-1] >= 1.0:
    if f1[-1] >= 0.9999:
        print 'Final value of f1: ',f1[-1]
        print 'Chosen value of f2[0]: ', f2[0]
        print 'Number of guesses: ',i
        break
```

```python
#Plot blasius solution and numerical solution

plt.figure(figsize=(14, 8))

# plotting x-component of velocity
plt.plot(f1,n,label="Blasius solution")
#plt.plot(f1*U,n,label="Blasius solution")
plt.xlabel('x-component of velocity',fontsize=18)

plt.plot(data0[:,1],eta,'-o',label='OpenFOAM')
plt.ylim(0,8)
plt.ylabel('$\eta$',fontsize=18)

plt.grid()
plt.legend(loc=0)
```

```python
#Plot blasius solution and numerical solution

plt.figure(figsize=(14, 8))

# plotting y-component of velocity
plt.plot((0.5*(f1*n-f)),n,label="Blasius solution")
plt.xlabel('y-component of velocity',fontsize=18)

plt.plot(data0[:,2]*rex**0.5,eta,'-o',label='OpenFOAM')

plt.ylim(0,8)
plt.ylabel('$\eta$',fontsize=18)

plt.grid()
plt.legend(loc=0)
```

```python
#Compute reynolds with plate distance

plate_lenght=2.0
x_plate=np.linspace(0,plate_lenght,1000)
re_x = U*x_plate/nu
small=1e-12
cf_re=0.664/(re_x+small)**(0.5)
```

```python
#Compute cf suing paraview data

ws_pv=np.sqrt(data2[:,7]**2.+data2[:,8]**2.+data2[:,9]**2.)
l_pv=data2[:,13]
cf_pv=ws_pv/(0.5*rho*U**2.)
```

```python
#Plot blasius solution and numerical solution of cf

plt.figure(figsize=(14, 8))

plt.plot(data1[:,0],cf_of,'-o', markevery=30,label="Sampling using OpenFOAM")
#plt.plot(data1[:,0],cf_of,'-',label="OpenFOAM")
plt.plot(x_plate,cf_re,label='Blasius solution')

plt.plot(l_pv,cf_pv,label='Sampling using paraView')

plt.xlabel('Plate length',fontsize=18)
plt.ylabel('Friction coefficient $c_f$',fontsize=18)

plt.xlim(0,2)
plt.ylim(0,0.015)
plt.grid()

plt.legend(loc=0)
```
