```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Compute therorical profiles
x=np.linspace(0.1,100000,1000000)
yp=x

#Laminar
up=yp
#Low-law
up1=(1.0/0.41)*np.log(yp)+5.0
```

```python
#Plot law of wall

#Plot profiles
plt.figure(figsize=(14, 8))

plt.plot(yp,up,label='Viscous sublayer')
plt.plot(yp,up1,label='Log-law')

plt.xscale('log')
#plt.xlim(0.1,100000)
plt.ylim(0,40)

plt.grid()
plt.legend(loc=2)
```

```python
#load cfl3d data at x=1.90334
data = np.loadtxt("CFL3D_X2.txt",skiprows=2)

#Antilogarithm of input data
al=10**data[:,0]
```

```python
#Load openfoam data
data1 = np.loadtxt("../postProcessing/sampleDict/2000/profile0_U_wallShearStress.xy",skiprows=0)
data2 = np.loadtxt("../postProcessing/sampleDict/2000/profile1_U_wallShearStress.xy",skiprows=0)

data3 = np.loadtxt("no_omega_WF/profile0_U_wallShearStress.xy",skiprows=0)
data4 = np.loadtxt("no_omega_WF/profile1_U_wallShearStress.xy",skiprows=0)
```

```python
#interpolate data
from scipy import interpolate
fint=interpolate.interp1d(data2[:,0],data2[:,4],kind='linear')
fint1=interpolate.interp1d(data4[:,0],data4[:,4],kind='linear')

#find value at a given point
#fint(xloc)
```

```python
#Where to sample
xloc=1.90334   

#Input values
U = 1
nu = 2.0e-07
rho = 1.0
```

```python
#Compute wall shear stress magnitude and velocity magnitude
#Shear stress units is in pascal must multiply by density in incompressible flows

ws = np.sqrt(data1[:,4]**2 + data1[:,5]**2 + data1[:,6]**2)
um = np.sqrt(data1[:,1]**2 + data1[:,2]**2 + data1[:,3]**2)

#wsm=0.0012160720369630369
wsm=np.abs(fint(xloc))

utau=np.sqrt(wsm/rho)
ypn=utau*data1[:,0]/nu
upn=um/utau
```

```python
#Compute wall shear stress magnitude and velocity magnitude
#Shear stress units is in pascal must multiply by density in incompressible flows

ws1 = np.sqrt(data3[:,4]**2 + data3[:,5]**2 + data3[:,6]**2)
um1 = np.sqrt(data3[:,1]**2 + data3[:,2]**2 + data3[:,3]**2)

#wsm1=0.0012272871987512486
wsm1=np.abs(fint1(xloc))

utau1=np.sqrt(wsm1/rho)
ypn1=utau1*data3[:,0]/nu
upn1=um1/utau1
```

```python
#Plot profiles

plt.figure(figsize=(14, 8))

#Correlations
plt.plot(yp,up,label='Viscous sublayer')
plt.plot(yp,up1,label='log-law')

#cfl3d
#plt.plot(ypn,upn,'-o',ms=6)
plt.plot(al,data[:,1],'-',label='CFL3D')

#OpenFOAM
#plt.plot(ypn[0:],upn[0:],'-o',label='OpenFOAM')
plt.plot(ypn[1:],upn[1:],'-o', ms=5, label='OpenFOAM V1 - Omega WF')

#In V2 omega wall function is switch or set a fixedValue
plt.plot(ypn1[1:],upn1[1:],'-o', ms=5, label='OpenFOAM V2 - Omega WF Disabled')



plt.xscale('log')
#plt.xlim(0.1,100000)
plt.ylim(0,40)

plt.grid()
plt.legend(loc=2)
```
