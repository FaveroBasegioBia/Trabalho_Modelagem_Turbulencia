```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load openfoam data
ke1 = np.loadtxt("ke/sampleDict/2000/profile0_U_wallShearStress.xy",skiprows=0)
ke2 = np.loadtxt("ke/sampleDict/2000/profile1_U_wallShearStress.xy",skiprows=0)

kwhre1 = np.loadtxt("kwhre/sampleDict/2000/profile0_U_wallShearStress.xy",skiprows=0)
kwhre2 = np.loadtxt("kwhre/sampleDict/2000/profile1_U_wallShearStress.xy",skiprows=0)

kwlre1 = np.loadtxt("kwlre/sampleDict/2000/profile0_U_wallShearStress.xy",skiprows=0)
kwlre2 = np.loadtxt("kwlre/sampleDict/2000/profile1_U_wallShearStress.xy",skiprows=0)

laminar1 = np.loadtxt("laminar/sampleDict/2000/profile0_U_wallShearStress.xy",skiprows=0)
laminar2 = np.loadtxt("laminar/sampleDict/2000/profile1_U_wallShearStress.xy",skiprows=0)

sa1 = np.loadtxt("sa/sampleDict/2000/profile0_U_wallShearStress.xy",skiprows=0)
sa2 = np.loadtxt("sa/sampleDict/2000/profile1_U_wallShearStress.xy",skiprows=0)
```

```python
#Compute therorical profiles
x=np.linspace(0.1,100000,1000000)
yp=x

#Laminar
up=yp
#Low-law
up1=(1.0/0.41)*np.log(yp)+5.0
```

```python
#interpolate data
from scipy import interpolate
fint1=interpolate.interp1d(kwhre2[:,0],kwhre2[:,4],kind='linear')
fint2=interpolate.interp1d(kwlre2[:,0],kwlre2[:,4],kind='linear')
fint3=interpolate.interp1d(sa2[:,0],sa2[:,4],kind='linear')
fint4=interpolate.interp1d(laminar2[:,0],laminar2[:,4],kind='linear')
fint5=interpolate.interp1d(ke2[:,0],ke2[:,4],kind='linear')

#find value at a given point
#fint(xloc)
```

```python
#Where to sample
xloc=1.90334   

#Input values
U = 1
nu = 2.0e-07
rho = 1.0
```

```python
#Compute wall shear stress magnitude and velocity magnitude
#Shear stress units is in pascal must multiply by density in incompressible flows

#KW HIGH RE
ws1 = np.sqrt(kwhre1[:,4]**2 + kwhre1[:,5]**2 + kwhre1[:,6]**2)
um1 = np.sqrt(kwhre1[:,1]**2 + kwhre1[:,2]**2 + kwhre1[:,3]**2)

#wsm=0.0012264945506493506
wsm1=np.abs(fint1(xloc))

utau1=np.sqrt(wsm1/rho)
ypn1=utau1*kwhre1[:,0]/nu
upn1=um1/utau1


#KW LOW RE
ws2 = np.sqrt(kwlre1[:,4]**2 + kwlre1[:,5]**2 + kwlre1[:,6]**2)
um2 = np.sqrt(kwlre1[:,1]**2 + kwlre1[:,2]**2 + kwlre1[:,3]**2)

#wsm=0.0012264945506493506
wsm2=np.abs(fint2(xloc))

utau2=np.sqrt(wsm2/rho)
ypn2=utau2*kwlre1[:,0]/nu
upn2=um2/utau2


#SA
ws3 = np.sqrt(sa1[:,4]**2 + sa1[:,5]**2 + sa1[:,6]**2)
um3 = np.sqrt(sa1[:,1]**2 + sa1[:,2]**2 + sa1[:,3]**2)

#wsm=0.0012264945506493506
wsm3=np.abs(fint3(xloc))

utau3=np.sqrt(wsm3/rho)
ypn3=utau3*sa1[:,0]/nu
upn3=um3/utau3


#laminar
ws4 = np.sqrt(laminar1[:,4]**2 + laminar1[:,5]**2 + laminar1[:,6]**2)
um4 = np.sqrt(laminar1[:,1]**2 + laminar1[:,2]**2 + laminar1[:,3]**2)

#wsm=0.0012264945506493506
wsm4=np.abs(fint4(xloc))

utau4=np.sqrt(wsm4/rho)
ypn4=utau4*laminar1[:,0]/nu
upn4=um4/utau4


#kepsilon
ws5 = np.sqrt(ke1[:,4]**2 + ke1[:,5]**2 + ke1[:,6]**2)
um5 = np.sqrt(ke1[:,1]**2 + ke1[:,2]**2 + ke1[:,3]**2)

#wsm=0.0012264945506493506
wsm5=np.abs(fint5(xloc))

utau5=np.sqrt(wsm5/rho)
ypn5=utau5*ke1[:,0]/nu
upn5=um5/utau5
```

```python
#Plot profiles

plt.figure(figsize=(14, 8))

#Correlations
plt.plot(yp,up,label='Viscous sublayer')
plt.plot(yp,up1,label='log-law')

#OpenFOAM
plt.plot(ypn1[1:],upn1[1:],'-o',ms=4,label='$\kappa$-$\omega$ High RE')
plt.plot(ypn2[0:],upn2[0:],'-',label='$\kappa$-$\omega$ Low RE')
plt.plot(ypn3[0:],upn3[0:],'-',label='Spalart-Allmaras')
plt.plot(ypn4[0:],upn4[0:],'o',ms=4,label='Laminar')
plt.plot(ypn5[1:],upn5[1:],'-o',ms=3,c='brown',label='$\kappa$-$\epsilon$')

plt.xscale('log')
#plt.xlim(0.1,1000000)
plt.ylim(0,40)

plt.grid()
plt.legend(loc=2)

#plt.savefig("figbl.png")
```

```python
#Compute wall shear stress magnitude and velocity magnitude
#Shear stress units is in pascal must multiply by density in incompressible flows

ws1 = np.sqrt(kwhre2[:,4]**2 + kwhre2[:,5]**2 + kwhre2[:,6]**2)
ws2 = np.sqrt(kwlre2[:,4]**2 + kwlre2[:,5]**2 + kwlre2[:,6]**2)
ws3 = np.sqrt(sa2[:,4]**2 + sa2[:,5]**2 + sa2[:,6]**2)
ws4 = np.sqrt(laminar2[:,4]**2 + laminar2[:,5]**2 + laminar2[:,6]**2)
ws5 = np.sqrt(ke2[:,4]**2 + ke2[:,5]**2 + ke2[:,6]**2)
```

```python
#Compute plate reaynolds X and friction coefficient

x_plate=np.linspace(0.00001,2,1000)
re_x=x_plate*U/nu
small=1e-12
#cf_re=0.0592/(re_x+1*small)**(0.2)
cf_re=0.0592/(re_x)**(0.2)

cf_of1=ws1/(0.5*rho*U**2)
cf_of2=ws2/(0.5*rho*U**2)
cf_of3=ws3/(0.5*rho*U**2)
cf_of4=ws4/(0.5*rho*U**2)
cf_of5=ws5/(0.5*rho*U**2)
```

```python
#Plot data

plt.figure(figsize=(14, 8))

plt.plot(kwhre2[:,0],cf_of1,'-',label='$\kappa$-$\omega$ High RE')
plt.plot(kwlre2[:,0],cf_of2,'-',label='$\kappa$-$\omega$ Low RE')
plt.plot(sa2[:,0],cf_of3,'-',label='Spalart-Allmaras')
plt.plot(laminar2[:,0],cf_of4,'-',label='Laminar')
plt.plot(ke2[:,0],cf_of5,'-',label='$\kappa$-$\epsilon$ ')

plt.xlabel('Plate length',fontsize=18)
plt.ylabel('Friction coefficient $c_f$',fontsize=18)

plt.xlim(0,2)
plt.ylim(0,0.015)
plt.grid()

plt.legend(loc=0)

#plt.savefig("figcf.png")
```
